## Some comments:

"the Quantum Fourier Transform is the classical Discrete Fourier Transform mapped onto the orthonormal basis of a finite-dimensional complex Hilbert space."

"In short, QFT is a change of basis from the (quantum) computational basis to the Fourier basis."
